# Spotify Exploratory Data Analysis

 The objective of this data analysis project for a Spotify dataset is to explore and analyze a rich collection of music data that includes audio features, such as tempo, loudness, danceability, and valence, as well as track details like popularity, release date, and artist information.

By leveraging various data analysis techniques, we aim to gain insights into the listening behavior of Spotify users, genre preferences, and the factors that contribute to the success of songs and artists on the platform. Through the use of statistical analysis, we can identify trends and patterns within the data, such as the most popular genres, the correlation between audio features and song popularity, and the characteristics of highly streamed songs.

In addition, we will utilize various data visualization techniques to effectively communicate our findings and insights. For example, we may create bar charts or scatter plots to showcase the relationship between audio features and song popularity or create a network graph to visualize the most influential artists on the platform.

Overall, this data analysis project for a Spotify dataset provides an opportunity to explore a large collection of music data and gain valuable insights into user preferences and trends, which can be leveraged to inform recommendations for users, artists, and the music industry.





 <b>Import libraries used</b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
import seaborn as sns
```

Import the CSV file 

```python
df_tracks = pd.read_csv("tracks.csv")
df_tracks.head()
```

Check number of null values in every column

```python
df_tracks.isnull().sum()
```

 Check number of rows and columns in the dataset

```python
df_tracks.shape
```

 Sort the data in descending order of popularity to find the most popular songs

```python
sorted_df = df_tracks.sort_values("popularity",ascending = False)
sorted_df.head()
```

 Calculate various statistics of the data

```python
df_tracks.describe().transpose()
```

 Find information about the songs whose popularity is greater than 90

```python
most_popular= df_tracks[df_tracks["popularity"]>90].sort_values("popularity",ascending=False)
most_popular.head()
```

 Extract a particular artist using indexing

```python
df_tracks[["artists"]].iloc[18]
```

 Convert the duration of songs from milliseconds to seconds

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
```

 Drop the milliseconds duration column from the data

```python
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
df_tracks.head()
```

```python
df_tracks["artists"].value_counts()
```

 Correaltion heatmap between various variables in the dataset

```python
corr_df = df_tracks.drop(["key","mode","explicit"],axis = 1).corr(method="pearson")
py.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot= True)
heatmap.set_title("Correlation heatmap between the variables")                   
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
sample_df.head()
```

```python
len(sample_df)
```

Scatter plot to visualize relationship between loudness and energy

```python
py.figure(figsize=(14,8))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="Loudness v/s Energy")
```

Scatter plot to visualize relationship between popularity and acousticness

```python
py.figure(figsize=(14,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="c").set(title="Popularity v/s Acousticness")
```

```python
df_tracks["Dates"]=pd.to_datetime(df_tracks["release_date"])
df_tracks.head()
```

```python
df_tracks["Year"] = df_tracks["Dates"].dt.year
df_tracks.head()
```

Histogram to visuaize no. of songs per year

```python
sns.displot(df_tracks["Year"],discrete= True, aspect=2,height=5,kind="hist").set(title="No. of songs per year")
```

```python
total_dr = df_tracks["duration"]
```

Line plot to visualize relationship between year and duration of songs

```python
sns.set_style(style="whitegrid")
fig,ax=py.subplots(figsize=(14,8))
fig=sns.lineplot(x=df_tracks["Year"],y=total_dr,ax=ax).set(title="year vs duration")
```

```python
df = pd.read_csv("SpotifyFeatures.csv")
df.head()
```

Bar graph to visualize relationship between duration and genre

```python
py.title("Duration of songs in different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y=df["genre"],x=df["duration_ms"],data=df)
py.xlabel("Duration(ms)")
py.ylabel("Genres")
```

Bar graph to visualize top genres on the basis of popularity

```python
sns.set_style(style="whitegrid")
py.figure(figsize=(14,8))
famous = df.sort_values("popularity",ascending=False).head(10)
sns.barplot(x="genre",y="popularity",data=famous).set(title="Top genres by popularity")
```
